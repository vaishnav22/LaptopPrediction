```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv('LaptopPricePrediction.csv')
```

```python
df.head()
```

# cleaning the dataset

```python
#lets reomove the fist unnamed column
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
#cleaning RAM column
```

```python
df['RAM'] = df.RAM.str.replace('\s.*','')
```

```python
df.head()
```

```python
#cleaning Storage
```

```python
df['Storage'] = df.Storage.replace('1 TB','1024 GB',regex=True)
```

```python
df.head()
```

```python
#display has data in cm and inches we can keep either one of them
df['Display'] = df.Display.str.replace('\s.*','',regex=True)
```

```python
df.head()
```

```python
#cleaning the price column
```

```python
price = df.Price.apply(lambda x:x.split('₹')[1])
```

```python
df.Price = price
```

```python
df.head(3)
```

```python
#processing warranty column
df.Warranty = df.Warranty.str.replace('\s.*','',regex=True)
```

```python
df.drop(['rating', 'Operating System'],axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df["Price"] = df["Price"].astype(str).astype(int)
print(df.dtypes)
```

```python
column = [Predictor for Predictor in df.columns if df[Predictor].dtype == 'O']
```

```python
for Predictor in column:
    labels_ordered= df.groupby([Predictor])['Price'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[Predictor]=df[Predictor].map(labels_ordered)
df.head(5)
```

```python
#seprating target vraible from features
Target = df['Price']
features = df.drop(columns='Price')
```

# scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 

a = scaler.fit_transform(features)
features = pd.DataFrame(a)
```

# spliting the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, Target, test_size=0.3, random_state=42,shuffle = True)
```

```python
from sklearn.ensemble import RandomForestRegressor
Random_forest_regressor = RandomForestRegressor(random_state=10)
Random_forest_regressor.fit(X_train,y_train)
```

```python
print("Training score")
print(Random_forest_regressor.score(X_train,y_train))
print('\r')
print("Testing score")
print(Random_forest_regressor.score(X_test,y_test))
```
